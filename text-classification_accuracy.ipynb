```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline

tokenizer = AutoTokenizer.from_pretrained("valhalla/distilbart-mnli-12-3")
model = AutoModelForSequenceClassification.from_pretrained("valhalla/distilbart-mnli-12-3")
sequence_to_classify = ('Joe waited for the train'),
                    
classifier = pipeline("zero-shot-classification",tokenizer=tokenizer,model=model)
# we will classify the Russian translation of, "Who are you voting for in 2020?"

# we can specify candidate labels in Russian or any other language above:
candidate_labels = ["food", "vehicle", "politics"]


classifier(sequence_to_classify, candidate_labels)

```
